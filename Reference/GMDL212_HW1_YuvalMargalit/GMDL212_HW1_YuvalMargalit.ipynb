```python
import numpy as np
import pylab
pylab.ion()
from pylab import plt
```

```python
x2,x1 = np.mgrid[-3:3:0.01,-3:3:0.01]
```

```python
def QnormSquared(Q):
    return x1**2*Q[0,0]+x2**2*Q[1,1]+2*x1*x2*Q[0,1]
```

```python
def axis_xy(g=None):
    # mimics Matlab's "axis xy" command;
    # making the y axis pointing upward when using imshow
    if g is None:
        g = plt.gca()
    bottom, top = g.get_ylim()  
    if top<bottom:
        g.set_ylim(top,bottom)
```

```python
def visualize(Q):
    plt.figure()
    plt.imshow((QnormSquared(Q)<1).astype(np.int64),cmap='gray')
    axis_xy()
    plt.colorbar()
    plt.axis('off')
    plt.title(r'$x^TQx<1$')
    plt.subplots_adjust(wspace=0.4)
    
```

```python
Q=np.eye(2)
print ('Q=\n',Q)
visualize(Q)
```

```python
plt.figure()
plt.imshow((abs(x1)+abs(x2)<1).astype(np.int64),cmap='gray')
axis_xy()
plt.colorbar()
plt.axis('off')
plt.title(r'L1')
plt.subplots_adjust(wspace=0.4)
```

```python
Q=np.eye(2)*np.diag([9,1])
print ('Q=\n',Q)
visualize(Q)
```

```python
Q=np.array([[9,2],[2,1]])
print ('Q=\n',Q)
visualize(Q)
```

```python
Q=np.array([[9,-2],[-2,1]])
print ('Q=\n',Q)
visualize(Q)
```
